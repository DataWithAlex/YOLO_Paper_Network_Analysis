```python
%pip install networkx pandas

import networkx as nx
import pandas as pd

# Load the .graphml file
graph = nx.read_graphml('backward_citation_network_with_extended_metadata.graphml')

# Convert node data to pandas DataFrame
# This will create a DataFrame where each row represents a node and its attributes
nodes_df = pd.DataFrame.from_dict(dict(graph.nodes(data=True)), orient='index')

# Convert edge data to pandas DataFrame
# This will create a DataFrame where each row represents an edge and its attributes
edges_df = pd.DataFrame.from_dict(nx.to_pandas_edgelist(graph))

# Display the DataFrames
print("Nodes DataFrame:")
print(nodes_df.head())
print("\nEdges DataFrame:")
print(edges_df.head())
```
